```python
import folium


edge_file = 'data_folium/EdgeTrainFile.geojson'
m_edge='data_folium/EdgeMetroFile.geojson'


my_map = folium.Map(location=[48.7,2.3] , zoom_start=10, width=800, height = 800)
```

```python
folium.GeoJson(open('data_folium/EdgeTrainFile.geojson'), name='geojson').add_to(my_map)

my_map.geo_json(geo_path=m_edge, line_weight=4)
#my_map.geo_json(geo_path=edge_file, line_weight=3)

folium.LayerControl().add_to(my_map)
```
